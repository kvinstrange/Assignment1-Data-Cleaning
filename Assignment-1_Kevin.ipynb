
# DATA CLEANING

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./Traffic_Crashes.csv')
```

```python
df.dropna(how="all")
```

```python
df.dropna(axis="columns", how="all")
```

```python
df.isna().sum()
```

```python
# Drop columns with high missing values
columns_to_drop = [
    'LANE_CNT', 'CRASH_DATE_EST_I', 'INTERSECTION_RELATED_I', 'NOT_RIGHT_OF_WAY_I', 'HIT_AND_RUN_I',
    'PHOTOS_TAKEN_I', 'STATEMENTS_TAKEN_I', 'DOORING_I', 'WORK_ZONE_I',
    'WORK_ZONE_TYPE', 'WORKERS_PRESENT_I','BEAT_OF_OCCURRENCE'
]

df = df.drop(columns=columns_to_drop)
```

```python
df
```

```python
duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {duplicates}")
```

```python
# Check for columns with missing values
missing_values = df.isna().sum()

missing_columns = missing_values[missing_values > 0]
missing_columns
```

```python
# Fill missing values for categorical variables
df['REPORT_TYPE'].fillna("Unknown", inplace=True)
df['STREET_NAME'].fillna("Unknown", inplace=True)
df['MOST_SEVERE_INJURY'].fillna("No Injury", inplace=True)  # "No Injury" (assuming no report means no injury).
```

```python
# Fill missing 'STREET_DIRECTION' with the most common value (mode)
df['STREET_DIRECTION'].fillna(df['STREET_DIRECTION'].mode()[0], inplace=True)

#Fill with 0 (assuming missing means no injury, cann't put mean mode without knowing the injury)
injury_columns = [
    'INJURIES_TOTAL', 'INJURIES_FATAL', 'INJURIES_INCAPACITATING', 
    'INJURIES_NON_INCAPACITATING', 'INJURIES_REPORTED_NOT_EVIDENT', 
    'INJURIES_NO_INDICATION', 'INJURIES_UNKNOWN'
]
df[injury_columns] = df[injury_columns].fillna(0)     
```

```python
df.dropna(subset=['LATITUDE', 'LONGITUDE'], inplace=True)
```

```python
# Verify if all missing values are handled
missing_values_after = df.isna().sum()
print(missing_values_after)
```

```python
# Convert column names to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# Convert Date/Time Columns
df['crash_date'] = pd.to_datetime(df['crash_date'], errors='coerce')
df['date_police_notified'] = pd.to_datetime(df['date_police_notified'], errors='coerce')
```

```python
df.head()
```

```python
# Standardize Categorical Values
df['weather_condition'] = df['weather_condition'].str.strip().str.lower()
df['traffic_control_device'] = df['traffic_control_device'].str.strip().str.lower()
df['lighting_condition'] = df['lighting_condition'].str.strip().str.lower()
df['street_name'] = df['street_name'].str.strip().str.lower()
```

```python
#Standardization for Key Columns

#Standardize Weather Condition Column**
df['weather_condition'] = df['weather_condition'].replace({
    'cloudy/overcast': 'cloudy',
    'fog/smoke/haze': 'fog',
    'freezing rain/drizzle': 'freezing_rain',
    'sleet/hail': 'sleet',
    'blowing snow': 'snow',
    'severe cross wind gate': 'windy',
    'blowing sand, soil, dirt': 'dust_storm'
})

#Standardize Street Names (Expand Abbreviations)**
df['street_name'] = df['street_name'].replace({
    'ave': 'avenue',
    'st': 'street',
    'rd': 'road',
    'blvd': 'boulevard',
    'dr': 'drive',
    'pl': 'place',
    'ln': 'lane',
    'hwy': 'highway',
    'pkwy': 'parkway',
    'ctr': 'center'
})

#Standardize Traffic Control Device Types**
df['traffic_control_device'] = df['traffic_control_device'].replace({
    'stop sign/flasher': 'stop_sign',
    'other reg. sign': 'other_regulatory_sign',
    'railroad crossing gate': 'railroad_gate',
    'rr crossing sign': 'railroad_sign',
    'flashing control signal': 'flashing_signal',
    'pedestrian crossing sign': 'pedestrian_sign',
    'school zone': 'school_zone',
    'other warning sign': 'warning_sign',
    'no passing': 'no_passing_zone'
})

# Verify the standardization
print(df['weather_condition'].unique())
print(df['street_name'].unique())
print(df['traffic_control_device'].unique())
```

```python
df
```

```python
# Find rows where DATE_POLICE_NOTIFIED is earlier than CRASH_DATE
invalid_dates = df[df['date_police_notified'] < df['crash_date']]
# Count and drop invalid rows
num_invalid_rows = len(invalid_dates)
df = df.drop(invalid_dates.index)
```

```python
# Create Crash Season Column
season_mapping = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 
                  6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall', 12: 'Winter'}
df['crash_season'] = df['crash_month'].map(season_mapping)
```

```python
df
```

```python
df['crash_year'] = df['crash_date'].dt.year
df['crash_month'] = df['crash_date'].dt.month
df['crash_hour'] = df['crash_date'].dt.hour
df['crash_day_of_week'] = df['crash_date'].dt.day_name()
```

```python
# Create Day/Night Column
df['day_night'] = df['crash_hour'].apply(lambda x: 'Day' if 6 <= x < 18 else 'Night')
```

```python
df = df[df['posted_speed_limit'].between(10, 70)]
```

```python
df.columns
```

```python
df.info()
```

```python
# Save dataset with new columns
df.to_csv(" clean_chicago_traffic.csv", index=False)
```

# CREATING CHARTS

```python
df = pd.read_csv("./ clean_chicago_traffic.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the Seaborn style and palette
sns.set_palette("rocket")

plt.figure(figsize=(12, 6))  # Make the figure larger
df['crash_year'].value_counts().sort_index().plot(kind='bar', color=sns.color_palette("rocket", n_colors=len(df['crash_year'].value_counts().sort_index())), width=0.8)

# Title and labels with custom font sizes
plt.title("Number of Crashes Per Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Crashes", fontsize=12)

# Adjust tick label sizes
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['crash_month'], palette="viridis")
plt.title("Number of Crashes Per Month", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Accident Count", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['crash_day_of_week'], 
              order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], 
              palette="mako")
plt.title("Accidents by Day of the Week", fontsize=14)
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Accident Count", fontsize=12)
plt.xticks(fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df['weather_condition'], order=df['weather_condition'].value_counts().index)
plt.title("Crashes by Weather Condition")
plt.xlabel("Number of Crashes")
plt.ylabel("Weather Condition")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by year and day/night to see trends over time
crash_trend = df.groupby(['crash_year', 'day_night']).size().unstack()

# Create an area plot
plt.figure(figsize=(14, 6))
crash_trend.plot(kind='area', stacked=True, alpha=0.6, colormap='coolwarm')

# Graph labels
plt.title("Crashes Over the Years: Day vs. Night")
plt.xlabel("Year")
plt.ylabel("Number of Crashes")
plt.legend(title="Time of Day")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
